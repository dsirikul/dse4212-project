# Import Packages

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
```

# Technology Stocks

```python
# List of Stocks
stock_tickers = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA']

# Initialise List to store RMSE and MAE values
rmse_values = []
mae_values = []

# Loop through each stock ticker
for stock_ticker in stock_tickers:
    print(f'Processing {stock_ticker}...')
    
   # Load the data with date parsing
    df = pd.read_csv(f'{stock_ticker}_processed_data.csv', parse_dates=['Date'])

    # Remove the first 49 rows due to NaN values in SMA columns
    df = df.iloc[49:]

    # Scaling the data
    scaler = MinMaxScaler(feature_range=(0, 1))

    scaled_data = scaler.fit_transform(df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume',
                                                 '5_day_SMA', '20_day_SMA', '50_day_SMA', '14_day_RSI',
                                                 'Lagged_Close', 'ATR', 'Day_of_Week', 'Daily_Return', 'Volatility']]) # Selecting only the wanted input features

    # Prepare the data for LSTM (Sequencing)
    look_back = 60  # No. of previous time steps to use

    X, y = [], []

    for i in range(look_back, len(scaled_data)):
        X.append(scaled_data[i - look_back : i])  # Create sequences of past data
        y.append(scaled_data[i, 3])  # Use the 'Close' price as the output, 3 being the location of 'Close'

    X, y = np.array(X), np.array(y)

    # Split the data into training and test sets
    split = int(0.8 * len(X)) # 80% Train 20% Validate

    Xtrain, Xtest = X[:split], X[split:]
    ytrain, ytest = y[:split], y[split:]

    # Getting the corresponding dates for the test set
    test_dates = df['Date'].iloc[look_back + split:].values

    # Building the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(Xtrain.shape[1], Xtrain.shape[2])))
    model.add(Dropout(0.2))  # Dropout layer to prevent overfitting

    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))

    model.add(Dense(units=1))  # Dense layer to produce the final prediction

    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fitting the model
    history = model.fit(Xtrain, ytrain, epochs=50, batch_size=32, validation_data=(Xtest, ytest), verbose=1)

    # Predictions
    predicted_prices = model.predict(Xtest)

    # Inverse transformation for predicted prices
    predicted_prices_reshaped = np.zeros((predicted_prices.shape[0], scaled_data.shape[1])) # Creating 2D arrays with the same shape as scaled_data
    predicted_prices_reshaped[:, 3] = predicted_prices.flatten()  # Fill only the 'Close' prices

    predicted_prices_inverse = scaler.inverse_transform(predicted_prices_reshaped)[:, 3]  # Get only the 'Close' prices

    # Inverse transformation for actual prices
    ytest_reshaped = np.zeros((ytest.shape[0], scaled_data.shape[1]))  # Creating 2D arrays with the same shape as scaled_data
    ytest_reshaped[:, 3] = ytest.flatten()  # Fill only the 'Close' prices

    actual_prices_inverse = scaler.inverse_transform(ytest_reshaped)[:, 3]  # Get only the 'Close' prices

    # Plotting the comparison of actual close prices with predicted close prices
    plt.figure(figsize=(14, 5))
    plt.plot(test_dates, actual_prices_inverse, color='blue', label='Actual Prices')
    plt.plot(test_dates, predicted_prices_inverse, color='red', label='Predicted Prices')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(f'{stock_ticker} Stock Price Prediction')
    plt.legend()
    plt.show()

    # Calculating the RMSE and MAE
    rmse = np.sqrt(mean_squared_error(actual_prices_inverse, predicted_prices_inverse))
    mae = mean_absolute_error(actual_prices_inverse, predicted_prices_inverse)

    print(f'{stock_ticker}_RMSE: {rmse}')
    print(f'{stock_ticker}_MAE: {mae}')

    # Append into the list for later mean calculation
    rmse_values.append(rmse)
    mae_values.append(mae)

# Calculating the mean RMSE and MAE across All stocks
mean_rmse = np.mean(rmse_values)
mean_mae = np.mean(mae_values)

print(f'Mean RMSE across Technology stocks: {mean_rmse}')
print(f'Mean MAE across Technology stocks: {mean_mae}')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, Dropout, MaxPooling1D
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

# Technology Stocks

```python
# List of Stocks
stock_tickers = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA']

# Initialise List to store RMSE and MAE values
rmse_values = []
mae_values = []

# Loop through each stock ticker
for stock_ticker in stock_tickers:
    print(f'Processing {stock_ticker}...')

    # Load and preprocess data
    df = pd.read_csv(f'Data/{stock_ticker}_processed_data.csv', parse_dates=['Date'])
    df = df.dropna()
    features = ['Open', 'High', 'Low', 'Adj Close', 'Volume', '5_day_SMA', '20_day_SMA', '50_day_SMA',
            '14_day_RSI', 'Lagged_Close', 'ATR', 'Day_of_Week', 'Daily_Return', 'Volatility']
    X = df[features].shift(1).dropna()
    y = df['Close'][1:].dropna()

    df = df.iloc[1:]  # Drop the first row of the original DataFrame
    dates = df['Date'].dropna()  # Get the adjusted dates

    # Scale features and reshape for CNN input
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
    X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

    # Split data into training and testing sets (80% train, 20% test)
    split = int(0.8 * len(X_scaled))
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y_scaled[:split], y_scaled[split:]
    dates_train, dates_test = dates[:split], dates[split:]

    # Define the CNN model
    model = Sequential([
        Conv1D(64, kernel_size=2, activation='relu', input_shape=(len(features), 1)),
        MaxPooling1D(pool_size=2),
        Dropout(0.3),
        Flatten(),
        Dense(50, activation='relu'),
        Dropout(0.3),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')

    # Train the model
    model.fit(X_train, y_train, epochs=60, batch_size=32, validation_data=(X_test, y_test))

    # Predict and inverse transform
    predicted_prices = model.predict(X_test).flatten()
    cnn_pred_reshaped = np.zeros((len(predicted_prices), X.shape[1]))
    cnn_pred_reshaped[:, 3] = predicted_prices
    cnn_pred_prices = scaler.inverse_transform(cnn_pred_reshaped)[:, 3]

    y_test_reshaped = np.zeros((len(y_test), X.shape[1]))
    y_test_reshaped[:, 3] = y_test.flatten()
    actual_prices = scaler.inverse_transform(y_test_reshaped)[:, 3]

    # Plot the predictions against actual prices
    plt.figure(figsize=(14, 5))
    plt.plot(dates_test, actual_prices, color='blue', label='Actual Prices')
    plt.plot(dates_test, cnn_pred_prices, color='red', label='Predicted Prices')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(f'{stock_ticker} Stock Price Prediction')
    plt.legend()
    plt.show()

    # Calculate RMSE and MAE
    rmse = np.sqrt(mean_squared_error(actual_prices, cnn_pred_prices))
    mae = mean_absolute_error(actual_prices, cnn_pred_prices)
    print(f'{stock_ticker}_RMSE: {rmse}, {stock_ticker}_MAE: {mae}')

    # Append the results for later calculation
    rmse_values.append(rmse)
    mae_values.append(mae)

# Calculating the mean RMSE and MAE across All stocks
mean_rmse = np.mean(rmse_values)
mean_mae = np.mean(mae_values)

print(f'Mean RMSE across Technology stocks: {mean_rmse}')
print(f'Mean MAE across Technology stocks: {mean_mae}')
```

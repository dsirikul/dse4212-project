```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define tickers for technology and healthcare stocks
tech_tickers = ["AAPL", "MSFT", "GOOG", "AMZN", "META", "NVDA"]
healthcare_tickers = ["LLY", "ABBV", "JNJ", "MRK", "TMO", "UNH"]

# Define start and end dates for data
start_date = "2014-09-01"
end_date = "2024-09-01"

# Download data for technology stocks
tech_data = {ticker: yf.download(ticker, start=start_date, end=end_date)["Close"] for ticker in tech_tickers}
tech_df = pd.DataFrame(tech_data)

# Download data for healthcare stocks
health_data = {ticker: yf.download(ticker, start=start_date, end=end_date)["Close"] for ticker in healthcare_tickers}
health_df = pd.DataFrame(health_data)

# Calculate sector averages
tech_df['Tech_Avg'] = tech_df.mean(axis=1)
health_df['Health_Avg'] = health_df.mean(axis=1)

# Define a list of colors for individual stocks
tech_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
health_colors = ['#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#1f77b4', '#ff7f0e']

# Plot for Technology Sector
plt.figure(figsize=(12, 6))
for i, col in enumerate(tech_df.columns[:-1]):  # Skip the 'Tech_Avg' column
    plt.plot(tech_df.index, tech_df[col], color=tech_colors[i % len(tech_colors)], linewidth=1, label=col)
plt.plot(tech_df.index, tech_df['Tech_Avg'], color='black', label='Technology Sector Average', linewidth=2.5)
plt.title('Technology Sector Stocks with Sector Average')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc='upper left')
plt.savefig("technology_sector_plot.jpg", format='jpg', dpi=300)  # Save as JPG
plt.show()

# Plot for Healthcare Sector
plt.figure(figsize=(12, 6))
for i, col in enumerate(health_df.columns[:-1]):  # Skip the 'Health_Avg' column
    plt.plot(health_df.index, health_df[col], color=health_colors[i % len(health_colors)], linewidth=1, label=col)
plt.plot(health_df.index, health_df['Health_Avg'], color='black', label='Healthcare Sector Average', linewidth=2.5)
plt.title('Healthcare Sector Stocks with Sector Average')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc='upper left')
plt.savefig("healthcare_sector_plot.jpg", format='jpg', dpi=300)  # Save as JPG
plt.show()
```
